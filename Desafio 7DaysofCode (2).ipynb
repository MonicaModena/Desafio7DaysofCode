
## Dia 2 - Manipulação dos dados

```python
# função para retornar a descrição da Classificação da Descrição Universal conforme código da localização
def classificar_CDU(codigo):
    if codigo <= 99:
        return 'Generalidades. Ciência e conhecimento'
    elif codigo <= 199:
        return 'Filosofia e psicologia'
    elif codigo <=  299:
        return 'Religião'
    elif codigo <=  399:
        return 'Ciências Sociais'
    elif codigo <= 499:
        return 'Classe vaga. Provisoriamente não ocupada'
    elif codigo <= 599:
        return 'Matemática e ciências naturais'
    elif codigo <= 699:
        return 'Ciências aplicadas'
    elif codigo <= 799:
        return 'Belas artes'
    elif codigo <= 899:
        return 'Linguagem. Língua. Linguística'
    else:
        return 'Geografia. Biografia. História'
```

```python
# aplicação da função para criar coluna
base_final['CDU'] = base_final['localizacao'].apply(classificar_CDU)
base_final
```

```python
base_final['CDU'].value_counts()
```

```python
# exclusão da coluna registro sistema, pois não faz sentido para a análise
base_final = base_final.drop("registro_sistema", axis = 1)
base_final
```

```python
# convertendo coluna Matricula ou siape em string
base_final['matricula_ou_siape'] = base_final['matricula_ou_siape'].astype(str)
base_final
```

```python
#Gráfico com exemplares emprestados por mês
sns.lineplot(data=qtde_exemplares_emprestados_mes, x='data_emprestimo', y='count', color='blue').set_title(
    'Exemplares emprestados por mês')
plt.xticks(rotation=30)
```

```python

def diferenca_percentual_ano_anterior(x,y):
  return round(((x / y * 100) - 100),2)
```

```python
percentual_2018 = diferenca_percentual_ano_anterior(tabela_pos.iloc[:,1],tabela_pos.iloc[:,0])
percentual_2019 = diferenca_percentual_ano_anterior(tabela_pos.iloc[:,2],tabela_pos.iloc[:,1])
percentual_2022 = diferenca_percentual_ano_anterior(tabela_pos.iloc[:,3],tabela_pos.iloc[:,2])
     
```

```python
#Cálculo do percentual de variação de empréstimos ao longo de 2018 a 2022
diferenca_percentual = pd.DataFrame({'2018':percentual_2018,
                           '2019':percentual_2019,
                           '2022':percentual_2022})
diferenca_percentual
```

```python
# parametrização dos dados para arquivo HTML
diferenca_percentual.reset_index(inplace=True)
diferenca_percentual.columns = diferenca_percentual.columns.str.capitalize()
diferenca_percentual.Curso = diferenca_percentual.Curso.str.capitalize()
```

```python
diferenca_percentual
```

```python

th_props = [
  ('font-size', '1.4rem'),
  ('text-align', 'center'),
  ('font-weight', 'bold'),
  ('color', 'whitesmoke'),
  ('background-color', '#001692'),
  ('border-radius', '0.25rem'),
  ('box-shadow','0 0 1rem gray')
  ]

td_props = [
  ('font-size', '1rem'),
  ('padding','0.5rem'),
  ('text-align', 'left'),
  ('font-weight', 'bold'),
  ('border-bottom','0.1rem solid lightgray')
  ]

styles = [
  dict(selector="th", props=th_props),
  dict(selector="td", props=td_props)
  ]
```

```python
diferenca_percentual.style.text_gradient(cmap='RdYlGn',low=1, axis=1,vmax=0.1,vmin=0)\
                              .format('{:.2f} %',subset=['2018','2019','2022'])\
                              .hide(axis="index")\
                              .set_table_styles(styles)\
                              .to_html('teste.html',doctype_html =True,
                                       table_attributes='ALIGN=LEFT WIDTH=50% CELLSPACING = 5')
```
