## Empréstimos dos acervos do sistema de bibliotecas da UFRN

### Importação empréstimos do acervo

```python
# Importando biblioteca Pandas
import pandas as pd
import glob
```

```python
# Importando base de dados de empréstimos dos 1º e 2º semestres dos anos de 2010 a 2020
lista_base = []
for arquivos in glob.glob(r"D:\Downloads\Desafio\*.csv"):
    lista_base.append(arquivos)
lista_base
```

```python
# Juntando as bases
listas = []
for arquivos in lista_base:
    listas.append(pd.read_csv(arquivos,index_col=0))
    
listas
```

```python

base_final_emprestimo = pd.concat(listas)
base_final_emprestimo
```

### Removendo duplicatas

```python
base_final_emprestimo = base_final_emprestimo.drop_duplicates()
base_final_emprestimo
```

### Tratando valores nulos

```python
# Identificando valores nulos
base_final_emprestimo.isnull().sum()
```

```python
# Excluí coluna por entender que não será relevante para a análise
base_final_emprestimo = base_final_emprestimo.drop("matricula_ou_siape", axis = 1)
base_final_emprestimo
```

```python
# Identificando os tipos dos dados
base_final_emprestimo.info()
```

```python
# Por entender que nem todos renovam o empréstimo, foi incluído o valor 0 na coluna data da renovação
base_final_emprestimo['data_renovacao'].fillna(0, inplace = True)
base_final_emprestimo
```

```python
# Por entender que alguns não devolvem o empréstimo ou ainda está no prazo de para devolução, foi incluído o valor 0 na coluna data da devolucao
base_final_emprestimo['data_devolucao'].fillna(0, inplace = True)
base_final_emprestimo
```

```python
# Verificando se sobraram valores nulos
base_final_emprestimo.isnull().sum()
```

### Importando base dos exemplares do acervo

```python
acervo = pd.read_parquet(r"D:\Downloads\Desafio\dados_exemplares.parquet")
acervo
```

```python
# Verificando se há valores nulos na base
acervo.isnull().sum()
```

```python
# Verificando se há valores duplicados na base
acervo = acervo.drop_duplicates()
acervo
```

### Mesclando as duas bases em um único Dataframe

```python
base_final = pd.merge(base_final_emprestimo, acervo, on = 'codigo_barras', how = 'inner')
base_final
```

```python
# Verificando se após a mesclagem das base foram gerados valores nulos
base_final.isnull().sum()
```
